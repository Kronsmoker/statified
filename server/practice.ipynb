```python
def total_points(points_list):
    total = 0
    
    for points in points_list:
        total += points
    
    return total

print(total_points([10, 20, 30]))
```

```python
def get_average(points_list):
    if len(points_list)==0:
        return 0
    
    total = total_points(points_list)
    return total / len(points_list)

print(get_average([10,20,30,90]))
```
